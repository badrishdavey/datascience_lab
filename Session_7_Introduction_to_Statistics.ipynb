# Introduction to Statistics

Statistics is derived from Italian words 'State' and 'Statista' which means 'The Government' and 'Useful facts / information'.  It is a branch of mathematics and deals with
- Collection of facts/information/data
- Arrangement, classification and tabulation of data
- Analysis and detailed study of the collected data
- Interpretation, conclusions and understanding of the data 
- Helps important business decision making

## Types of Statistics

- **Descriptive Statistics** - describes the data collated or collected for a specific business purpose.

- **Inferential Statistics** - involves detailed study, analysis and interpretation of data to draw conclusions based on Descriptive statistics.

## Population vs Sample

- **Population** - is the collection of objects/people which form the larger group of the analysis.
- **Sample** - is also the collection of objects/people which are usually a subset of the larger population or sometimes is the only size available for analysis.

## Types of Data

- **Qualitative Data** - this data suggests only the quality of the variable/attribute. Also called as Categorical data, this attribute is useful for classifying data into groups or classes. It is more applicable when there are limited or fewer values possible for that data. Qualitative data cannot be measured.
  - *Example*
    - Region – South, North, East, West
    - Gender – Male, Female
    - Nationality – Indian, American, Nigerian
  - ***Nominal Data*** - assigns a name as a value. This value is just a representation of which category the value belongs to. This is usually a character data but may sometimes be represented as a numeric identity. No further analysis is possible other than being used for classification / categorization. They have some kind of an identity.
    - Company name
    - Region
    - Name of a person
  - ***Ordinal Data*** - is also similar to nominal data, but ordinal data can be ranked in relation to other values in the same variable. Ordinal data values have some level indications within themselves.
    - Small, Medium, Large
    - Bad, Average, Good, Very Good, Excellent
    - Scale of 1 to 5 (where the numbers 1-5 represent only ranking)
- **Quantitative Data** - this is all kinds of information / data which can be quantified by assigning numeric values to them. These can be any number from 0 or negative numbers up to infinity.  In simple terms, data that can be measured is called quantitative data. The possibility of values for this attribute is unlimited unlike Qualitative data.
  - *Example*
    - Height
    - Weight
    - Salary
    - Temperature
  - ***Interval Scale Data*** - are quantitative data, and can be used for numerical analysis. They might have equal intervals between the values, or can also be continuous data. They do not have ZERO reference point, which means they can be positive or negative numbers. Usually, it would be possible for us to take the difference between any 2 values in this kind of data to get a meaningful result, but a division of values would not fetch a meaning result.
    - Age groups
    - Salary ranges
    - dates
  - ***Ratio Scale Data*** - is measureable data, and can be used for all kinds of econometrics. They have a zero reference point, which means they are usually positive data.

### Measures of Central Tendency

#### Mean

It is calculated as the total of all the values by the number of values or frequency.

$$\textbf{Mean} = \frac{\Sigma_{i=1}^{N} x_i}{N}$$

- Mean is extremely sensitive to outliers as even one or few values can influence the center value
- Mean is the most common measure of central tendency
- Mean does not take into calculation any missing values
- Zero is still considered as a value in the data range
- Mean uses all the data and each value influences the mean
- Average is also similar to mean, but one major difference is that Average takes all missing values also into account, which will not be accurate as required
- Mean for a population is represented by $\mu$ (mu), and a Sample mean is represented by $\bar{X}$ or $\bar{x}$ (X bar or x bar)

#### Mode

Most repeated value in the data series.

- Usually repeat values occur in the data series only if you have limited possible values in that data
- mode represents the most popular or preferred value
- Used mostly for categorical /nominal data and not much quantitative data
  - When data is highly repetitive, you go for the mode
- 2, 3, 6, 3, 3, 3, 2, 5 $\rightarrow$ 3
- 2, 3, 4, 2, 3, 3, 2, 5 $\rightarrow$ 2 and 3
  - Data is called bi-modal
  - Data can also be tri-modal, or other forms of multi-modal
- 2, 3, 4, 5, 6 $\rightarrow$ non-modal

````
def sample_from(X, size):
  X2 = np.random.choice(X, size)
  return(X2.mean())
````

````
sample_from(np.random.uniform(10, 20, 100000), 1000)
````

````
def many_means(X, size, number_of_means):
  return(np.array([sample_from(X, size) for i in range(number_of_means)]))
````

````
from scipy.stats import skew, kurtosis
def analysis(X):
  mu = X.mean()
  sigma = X.std()
  print('Mean: {:8.4f}, Standard Deviation: {:8.4f}, Skewness: {:8.4f}, Kurtosis: {:8.4f}, Median: {:8.4f}'.format(mu, sigma, skew(X), kurtosis(X), sorted(X)[int(len(X)/2)]))
  sns.distplot(X)
  plt.show()
  X_1std = np.sum([i < mu + sigma and i > mu - sigma for i in X])
  X_2std = np.sum([i < mu + 2*sigma and i > mu - 2*sigma for i in X])
  X_3std = np.sum([i < mu + 3*sigma and i > mu - 3*sigma for i in X])
  print('1 STD: {:8.4f}, 2 STD: {:8.4f}, 3 STD: {:8.4f}'.format(X_1std / len(X), X_2std / len(X), X_3std / len(X)))
````

### Min-Max Standardization

````
def minmax_normalize(X):
  return((X - X.min())/(X.max() - X.min()))
````
