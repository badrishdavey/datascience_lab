### Dimentionality Reduction

```
# PCA - Principal Component Analysis
# Takes in n features -> Transforms them in a linear manner (new feature 1 = old feature 1 * 0.05 + old feature 2 * 2.5 + ...)
# new feature are created in such a way that new feature 1 contains maximumm variation, new feature 2 contains second most variation etc
# If you started with 100 features... you now have new 100 features and from 100 features you may want to just use top 3 features as they might contain 99% of the variation
```

```
from yellowbrick.model_selection import ValidationCurve
from sklearn.linear_model import Lasso
```

Cross Validation of 5 for a data with 100 rows:

- gr1 = 0 to 19
- gr2 = 20 to 39
- gr3 = 40 to 59
- gr4 = 60 to 79
- gr5 = 80 to 99

Make 5 sets out these: (cross validations) 

- Training = gr1, gr2, gr3, gr4 | Testing = gr5
- Training = gr1, gr2, gr3, gr5 | Testing = gr4
- Training = gr1, gr2, gr4, gr5 | Testing = gr3
- Training = gr1, gr3, gr4, gr5 | Testing = gr2
- Training = gr2, gr3, gr4, gr5 | Testing = gr1

```
from sklearn.base import BaseEstimator, TransformerMixin
class ColTwoNumber(BaseEstimator, TransformerMixin):
  def __init__(self):
    pass
  
  def fit(self, X=None, y=None):
    return(self)
  
  def transform(self, X):
    X[:,1] = list(range(len(X)))
    return(X)
```

```
c = ColTwoNumber()
c.fit(X)
c.transform(X)
```

```
c.fit_transform(X)
```

```
p = Pipeline([
    ('MeanImputer', SimpleImputer(strategy='mean')),
    ('Normalizer', MinMaxScaler()),
    ('Column2', ColTwoNumber())
])
```
